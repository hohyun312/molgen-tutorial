# Define a Model

We extract node- and graph-level features using GCN. Then we compute logit values for each action.

- Stop: calculated based on graph features.
- AddNode: calculated based on graph features.
- AddEdge: calculated based on two node features to be connected (similar to link prediction).

```python

class Policy(nn.Module):
    def __init__(self, env, graph_nn, activation="leaky_relu"):
        super().__init__()
        self.env = env
        self.graph_nn = graph_nn
                
        emb_dim = graph_nn.emb_dim
        self.stop_mlp = MLP([emb_dim, emb_dim, 1], activation)
        self.addnode_mlp = MLP([emb_dim, emb_dim, self.env.num_node_types], activation)
        self.addedge_mlp = MLP([emb_dim, emb_dim, emb_dim, self.env.num_edge_types], activation)
        self.activation = getattr(F, activation)
        
    def logits(self, g):
        x = self.graph_nn(g)            
        graph_features = gnn.global_add_pool(x, g.batch)
        
        i, j = g.non_edge_index
        non_edge_features = x[i] + x[j]
        
        stop_logits = self.stop_mlp(graph_features).flatten()
        addnode_logits = self.addnode_mlp(graph_features).flatten() 
        addedge_logits = self.addedge_mlp(non_edge_features).flatten()
        logits = [stop_logits, addnode_logits, addedge_logits]
        return torch.cat(logits, dim=0)
    
    def forward(self, states):
        g = self.collate(states)
        logits = self.logits(g)
        stop_batch = torch.arange(g.num_graphs)
        addnode_batch = torch.arange(g.num_graphs).repeat_interleave(self.env.num_node_types)
        addedge_batch = torch.arange(g.num_graphs).repeat_interleave(self.env.num_edge_types * g.num_non_edges)
        indices = torch.cat([stop_batch, addnode_batch, addedge_batch], dim=0)
        return ScatterCategorical(logits, indices)
    
    def collate(self, states):
        return gd.Batch.from_data_list(
            [self.env.to_Data(s) for s in states]
        )

        
        
# sample data
caffeine = "CN1C=NC2=C1C(=O)N(C(=O)N2C)C"
state = env.State.from_smiles(caffeine)
states, actions = env.get_bfs_trajectory(state)


# Model
# graph_nn = GPS(env, emb_dim=32, num_layers=3)

graph_nn = RGCN(env, emb_dim=32, num_layers=3)
model = Policy(env, graph_nn)
cat = model(states)
cat
```

# Sampler

```python
@torch.no_grad()
def sample_terminating_state(env, model, max_len=25):
    model.eval()
    
    state = env.initial_state()
    for _ in range(max_len):
        cat = model([state])
        action_idx = cat.sample().item()
        action = env.idx_to_Action(state, action_idx)
        state, done = env.step(state, action)

        if done:
            break
            
    return state

state = sample_terminating_state(env, model, max_len=8)
state.to_mol()
```

# REINFORCE


```python
def sample_states_and_actions(env, model, max_len=25):
    model.eval()
    
    states, actions = [], []
    
    state = env.initial_state()
    for _ in range(max_len):
        cat = model([state])
        action_idx = cat.sample().item()
        action = env.idx_to_Action(state, action_idx)
        next_state, done = env.step(state, action)
        
        states.append(state)
        actions.append(action)
        
        state = next_state
        if done:
            break
            
    return states, actions


def reward_func(state):
    reward = 0
    for u in state.node_types:
        if u == env.NodeType.O:
            reward += 0.5

    for e in state.edge_types:
        reward -= 0.2 * e.value 

    neighbor = [[] for _ in range(state.num_node)]
    for i, (u, v) in enumerate(state.edge_list):
        neighbor[u].append(v)
        neighbor[v].append(u)

    for nei in neighbor:
        if len(nei) == 0:
            reward -= 1
    
    return reward


```

```python
from tqdm import tqdm
import numpy as np


batch_size = 16
alpha = 5.


graph_nn = RGCN(env, emb_dim=32, num_layers=3)
model = Policy(env, graph_nn)
optimizer = torch.optim.Adam(model.parameters())


avg_rewards = []
pbar = tqdm(range(200))
for _ in pbar:
    
    losses, rewards = [], []
    for _ in range(batch_size):

        states, actions = sample_states_and_actions(env, model)
        action_indices = torch.LongTensor(
            [env.Action_to_idx(s, a) for s, a in zip(states, actions)]
        )
        log_prob = model(states).log_prob(action_indices).sum()
        reward = reward_func(states[-1])
        
        ent = - log_prob.exp() * log_prob

        loss = - log_prob * reward + alpha * ent

        losses.append(loss)
        rewards.append(reward)
        
    loss = torch.stack(losses).mean()
    

    optimizer.zero_grad()
    loss.backward()
    torch.nn.utils.clip_grad_norm_(
        model.parameters(),
        0.5
    )
    optimizer.step()
    
    avg_rewards.append(torch.tensor(rewards).mean().item())
    
    pbar.set_postfix({"reward": avg_rewards[-1]})
    
    
plt.plot(avg_rewards)
plt.show()
```

# Genetic Algorithm

```python
def mutate(model):
    model = deepcopy(model)
    for param in model.parameters():
        param.data = param + 0.02 * torch.randn_like(param)
    return model

@torch.no_grad()
def evaluate_model(model, n_samples=5):
    rewards = []
    for _ in range(n_samples):
        states, actions = sample_states_and_actions(env, model)
        action_indices = torch.LongTensor(
            [env.Action_to_idx(s, a) for s, a in zip(states, actions)]
        )
        log_prob = model(states).log_prob(action_indices).sum()
        reward = reward_func(states[-1])
        rewards.append(reward)
    return np.mean(rewards)
```

```python
from tqdm import tqdm
import numpy as np
from copy import deepcopy

def random_model():
    graph_nn = RGCN(env, emb_dim=32, num_layers=3)
    model = Policy(env, graph_nn)

    return model

model_pool = [random_model() for i in range(16)]
new_model_pool = model_pool


avg_rewards = []
pbar = tqdm(range(500))
for _ in pbar:
    model_pool = new_model_pool
    
    rewards = []
    for model in model_pool:
        reward = evaluate_model(model, n_samples=1)
        rewards.append(reward)

    topk = np.argsort(rewards)[-8:]

    new_model_pool = [model_pool[i] for i in topk] + [mutate(model_pool[i]) for i in topk]
            
    avg_rewards.append(np.mean(rewards))
    
    pbar.set_postfix({"reward": avg_rewards[-1]})


plt.plot(avg_rewards)
plt.show()
```
